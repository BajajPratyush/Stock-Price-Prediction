### LSTM are sensitive to the scale of the data.
LSTM model take into consideration the data that we insert that is the close column of the above dataset that we extracted and is currently stored in the df1 dataframe. Since the data is very large in number i.e. the data ranges from 100-300, and we know that LSTM is very sensitive to slightest of the changes to the data, it can give very unsure results if processes as it is. So to overcome this problem we scale the data from the range 0-1. 

We are using the sklearn library's MinMaxScaler. This library is used to scale the data. Since it requires the data in form of arrays we transform the data into an array form using the numpy library and then use the MinMaxScaler to change the scale of the data to fit 0-1.

# Step 2: Preprocessing the data
The first step before processing the data is to split the data into training data and test data. There are various methods for splitting the data and its cross validation i.e. hold out, hold one out, k-cross validation, bootstrapping, etc. For this project we are going with the hold out method because the other methods works very well when we have to do classification or we have a linear regression problem and this method we use is used when we have to deal with time series data i.e. data is ordered in time and we have to split them with their related time and cannot split randomly. In this method we simply split the data into two unequal parts. The bigger chunk of data is used for training the model and the rest of the data (smaller dataset is used to test the model).


After that we will preprocess the data.

```python
training_size=int(len(df1)*0.65)
test_size=len(df1)-training_size
train_data=df1[0:training_size,:]
test_data=df1[training_size:len(df1),:1]
```

### Preprocessing of the data and it's explanation
We know that the data we need to predict depends on the previous data i.e. the stock price today depends on the price of the stock of the past few weeks.
So we convert the data such that the data of x days lead to the price of the stock on x+1 day.
####                                                                                                    x-train
| Day 1 | Day 2 | Day 3 | ...... | Day 100 |
| --- | --- | --- | --- | --- |
| 120 | 130 | 135 | ...... | 200 |
| 130 | 135 | 140 | ...... | 205 |

#### y- train 
| Result |
|---|
|205|
|203|

In the above example we see that the data of the past 100 days leads to the result in y-train, i.e. that 120, 130, 135 ......, 200 resulted in 205 on 101th day and similarly we move the data ahead by one step and now 130, 135, 140, ..... , 205 resulted in 203 on 102th day.


Hence we preprocess the data in such a way for the training as well as the test data.

```python
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)
```

Since LSTM require the data to be in 3 dimensions i.e. it also requires a feature dimension, so we reshape our x-train and x-test data into 3-dimensions. The 3rd dimensions indicates the amout of feature we are building the model on. Since we are only testing on 1 feature we set the 3rd dimension as 1 for both x_train and x_test.

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
x_train =x_train.reshape(x_train.shape[0],x_train.shape[1] , 1)
x_test = x_test.reshape(x_test.shape[0],x_test.shape[1] , 1)
x_train.shape
```

```python
### Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

## Creating the LSTM Model
We are creating a sequential type of model in which every layer gives its output for the next layer to process further. We add three different LSTM layers where 50 represent the size of the output layer of that layer and return sequences mean that it will give output for the next layer. After that all the layers are later condensed into one layer using Dense and is compiled using mean squared error and loss can be seen during fitting the dataset to the model.

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32,verbose=1)
```

```python
train_predict = model.predict(x_train)
test_predict = model.predict(x_test)
```

```python
##Transformback to original form
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

After converting the data we use a step of 100 and for every value we go back a 100 days and use all the data as input to predict the next data and then store it in train predict plot. We repeat the same thing for test data and then predict them over the original data to compare the difference.

```python
### Plotting 
# shift train predictions for plotting
look_back=100
trainPredictPlot = numpy.empty_like(df1)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
# shift test predictions for plotting
testPredictPlot = numpy.empty_like(df1)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(df1)-1, :] = test_predict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(df1))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```
